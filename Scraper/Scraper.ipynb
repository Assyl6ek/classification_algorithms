```python
import io
import pandas as pd
from bs4 import BeautifulSoup
```

```python
txt = io.open("Teremok.txt", encoding='utf-8')
contents = txt.read()
soup = BeautifulSoup(contents, 'html.parser')
print(soup.prettify())
```

```python
positive = soup.find_all(class_="reviewPositive")
negative = soup.find_all(class_="reviewNegative")
```

```python
List = []
count = 0
for x in positive:
    r = x.find(class_="reviewText").get_text()
    List.append(r)
    count+=1
```

```python
statusListP = [1] * (count)
stat = pd.DataFrame(statusListP, columns=['Status'])
dfP = pd.DataFrame(List, columns=['Review'])
dfP = pd.concat([dfP, stat], axis=1)
dfP
```

```python
List = []
count = 0
for x in negative:
    r = x.find(class_="reviewText").get_text()
    List.append(r)
    count+=1
```

```python
statusListN = [0] * (count)
stat = pd.DataFrame(statusListN, columns=['Status'])
dfN = pd.DataFrame(List, columns=['Review'])
dfN = pd.concat([dfN, stat], axis=1)
dfN
```

```python
result = pd.concat([dfN, dfP], axis=0, ignore_index=True)
result
```

```python
result.to_csv('Reviews-GadkiyKoyot.csv', index=False)
```
